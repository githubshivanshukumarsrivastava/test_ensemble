```python
from sklearn.ensemble import BaggingClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score

```

```python
bagging_clf = BaggingClassifier(estimator=DecisionTreeClassifier(), n_estimators=50, random_state=42)
adaboost_clf = AdaBoostClassifier(estimator=DecisionTreeClassifier(), n_estimators=50, random_state=42)
```

```python
print("BaggingClassifier:")
print(f"Accuracy: {bagging_accuracy:.2f}")
print(f"Precision: {bagging_precision:.2f}")
print(f"Recall: {bagging_recall:.2f}")
```

```python
print("\nAdaBoostClassifier:")
print(f"Accuracy: {adaboost_accuracy:.2f}")
print(f"Precision: {adaboost_precision:.2f}")
print(f"Recall: {adaboost_recall:.2f}")
```

# task 8

```python
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report
```

```python
X_imbalanced, y_imbalanced = make_classification(n_classes=2, class_sep=2, 
                                                 weights=[0.9, 0.1], n_informative=3, 
                                                 n_redundant=1, flip_y=0, 
                                                 n_features=20, n_clusters_per_class=1, 
                                                 n_samples=1000, random_state=42)
```

```python
rf_clf = RandomForestClassifier(random_state=42)
rf_clf.fit(X_imbalanced, y_imbalanced)
```

```python
y_pred = rf_clf.predict(X_imbalanced)
```

```python
conf_matrix = confusion_matrix(y_imbalanced, y_pred)
class_report = classification_report(y_imbalanced, y_pred)
```

```python
print("Confusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(class_report)
```

```python

class Vehicle:
    def __init__(self, brand, color):
        self.brand = brand
        self.color = color

    def display_info(self):
        return  self.brand,self.color


class Car(Vehicle):
    def __init__(self, brand, color, model):
        super().__init__(brand, color)
        self.model = model

    def display_info(self):
        return self.brand, self.color,self.model


vehicle = Vehicle("Toyota", "Red")
print(vehicle.display_info())

car = Car("Honda", "Blue", "Civic")
print(car.display_info())
```

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.__marks = marks  

    
    def get_marks(self):
        return self.__marks

 
    def set_marks(self, marks):
        if 0 <= marks <= 100:  
            self.__marks = marks
        else:
            print("Invalid marks. Please enter a value between 0 and 100.")


student = Student("Alice", 85)
print(f"Student Name: {student.name}")
print(f"Marks: {student.get_marks()}")

student.set_marks(95) 
print(f"Updated Marks: {student.get_marks()}")

student.set_marks(110)  
```
